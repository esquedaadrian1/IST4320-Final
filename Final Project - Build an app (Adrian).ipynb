```python
import tkinter as tk
from tkinter import messagebox

class CalculatorApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Simple Calculator")
        self.root.geometry("300x400")
        self.root.resizable(False, False)

        # Initialize variables
        self.current_input = ""
        self.result_var = tk.StringVar()

        # Create UI
        self.create_ui()

        # Initialize history
        self.history = []

    def create_ui(self):
        # Result display
        display_frame = tk.Frame(self.root, height=50)
        display_frame.pack(fill=tk.X, padx=10, pady=10)

        result_entry = tk.Entry(
            display_frame,
            textvariable=self.result_var,
            font=('Arial', 18),
            bd=0,
            justify=tk.RIGHT,
            state='readonly'
        )
        result_entry.pack(fill=tk.X)

        # Button frame
        button_frame = tk.Frame(self.root)
        button_frame.pack(expand=True, fill=tk.BOTH, padx=10, pady=10)

        # Button layout
        buttons = [
            ('7', 0, 0), ('8', 0, 1), ('9', 0, 2), ('/', 0, 3),
            ('4', 1, 0), ('5', 1, 1), ('6', 1, 2), ('*', 1, 3),
            ('1', 2, 0), ('2', 2, 1), ('3', 2, 2), ('-', 2, 3),
            ('0', 3, 0), ('.', 3, 1), ('C', 3, 2), ('+', 3, 3),
            ('=', 4, 0, 3), ('H', 4, 3)  # H for History
        ]

        for button in buttons:
            text = button[0]
            row = button[1]
            col = button[2]
            colspan = button[3] if len(button) > 3 else 1

            btn = tk.Button(
                button_frame,
                text=text,
                font=('Arial', 14),
                command=lambda t=text: self.on_button_click(t)
            )
            btn.grid(
                row=row,
                column=col,
                columnspan=colspan,
                sticky="nsew",
                padx=2,
                pady=2,
                ipadx=10,
                ipady=10
            )

            # Configure row/column weights
            button_frame.grid_rowconfigure(row, weight=1)
            button_frame.grid_columnconfigure(col, weight=1)

    def on_button_click(self, button_text):
        if button_text == 'C':
            self.current_input = ""
            self.result_var.set("")
        elif button_text == '=':
            self.calculate_result()
        elif button_text == 'H':
            self.show_history()
        else:
            self.current_input += button_text
            self.result_var.set(self.current_input)

    def calculate_result(self):
        """Custom function to calculate the result"""
        if not self.current_input:
            return

        try:
            # Replace × with *, ÷ with / for eval
            expression = self.current_input.replace('×', '*').replace('÷', '/')
            result = eval(expression)

            # Add to history
            calculation = f"{self.current_input} = {result}"
            self.history.append(calculation)

            # Update display
            self.result_var.set(result)
            self.current_input = str(result)
        except ZeroDivisionError:
            messagebox.showerror("Error", "Cannot divide by zero!")
            self.current_input = ""
            self.result_var.set("")
        except Exception as e:
            messagebox.showerror("Error", "Invalid expression")
            self.current_input = ""
            self.result_var.set("")

    def show_history(self):
        """Custom function to show calculation history"""
        if not self.history:
            messagebox.showinfo("History", "No calculations yet!")
            return

        history_text = "\n".join(self.history)
        messagebox.showinfo("Calculation History", history_text)

if __name__ == "__main__":
    root = tk.Tk()
    app = CalculatorApp(root)
    root.mainloop()
```
